**Question 9** (6 points) How did the pandemic interact with long-term trends in library / media
consumption?  Using the data in the column `UsageClass` make a line graph plot that has two line trends, one for "DIGITAL" and one for "PHYSICAL". Comment on the trends you see in the graph.

```python
digital_df = spl_df.loc[spl_df['UsageClass'] == 'Digital'].groupby('CheckoutYear')[['Checkouts']].sum().reset_index()
physical_df = spl_df.loc[spl_df['UsageClass'] == 'Physical'].groupby('CheckoutYear')[['Checkouts']].sum().reset_index()

plt.plot(digital_df['CheckoutYear'], digital_df['Checkouts'], label = 'Digital')
plt.plot(physical_df['CheckoutYear'], physical_df['Checkouts'], label = 'Physical')
plt.legend()
plt.xlabel('Year')
plt.ylabel('Number of Checkouts')
plt.show()

digital_df
```

Total checkouts increased but there is no gap between digital and physiscal. There 
was not dramatic switch between the two as they stay equal throughout. There is a decrease
in checkouts after the pandemic
