## Sequences (built into Python)

Sequences are the most basic type of structure to group data. The most important are **lists** and **tuples**.

A **list** is an ordered sequence of values that can be changed - mutable objects (in basic python...no library needed)
- Create a list with square brackets "[]"

A **tuple** is an ordered sequence of values that **cannot** be changed - immutable objects
- Create a tuple with parentheses "()"

````python
#let's create a list of numbers
Passenger_age = [22, 38, 26, 35, 35]  #each element is separated by commas

Passenger_age
````

````python
#and a tuple
tuple_age = (22,38,26)
tuple_age
````

## We can also extract a "slice" of a list


The range of elements can be specified with colons. The output is a list starting at left index and stopping at (right index -1)


````python
Passengers = ['Mr. Owen Harris Braund','Mrs. John Bradley', 'Miss. Laina', 'Mrs. Jacques Heath'] # 4 elements
Passengers[1:3]# up to but not including the end of the slice
#The above prints out element 2 and element 3...index 1 and index 2
````

 - Half-closed interval \[start,end)


```python 
list[start: end]
```

### But wait....there's more

## Default Settings: List slicing

We can slice a list by 'leaving out' a starting (or ending) position. The default is to start at the beginning (or end respectively).

````python
Passengers[:3] #start at beginning stop at element end-1 -----> 3-1=2
````

There is an optional arugment we can use when slicing a list, the step.
In general 
```list[start: end: step]``` 
with defaults of step=1


````python
Passengers = ['Mr. Owen Harris Braund','Mrs. John Bradley', 'Miss. Laina', 'Mrs. Jacques Heath'] # 4 elements
Passengers[::2]
````
