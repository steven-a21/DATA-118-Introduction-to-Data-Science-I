### Types of Data

All values in a column should be both the same type and be comparable to each other in some way

* Numerical — Each value is from a fixed scale
  * Typically some type of measurement
  * Numerical ordering is meaningful

* Categorical — Each value is from a fixed inventory
  * May or may not have an ordering ([low, med, high] vs. [red, blue, green]
  * Categories are mutually exclusive

## Types of Data

**Numerical** — Each value is from a fixed scale
 - Can have *discrete* or *continuous* data
 

 
- Discrete: finite, countable - integers
    - age, number of kids
             
             

 
- Continuous: infinite, take on any real number in an interval
    - income, height, test scores
              
              

 
 - Numerical measurements are ordered
 - Differences are typically meaningful
    

    



When working with numerical data, it is usually most appropriate to visualize through 

- scatter plots
 
- line plots 
 
- histograms

### “Numerical” Values

Just because the values are numbers, doesn’t mean the variable is numerical

* Census data uses numerical value for `SEX` code (0, 1, and 2)

* It doesn’t make sense to perform arithmetic on these “numbers”, e.g. 1 - 0 or (0+1+2)/3 are nonsense here

* The variable `SEX` is still categorical, even though numbers were used for the categories

**Categorical** — Each value is from a fixed inventory
 - Can have *nominal*  or *ordinal* data
 

 
   - Nominal: a naming or labeling scale (that is NO natural order is present)
        - hair color, names of people
                 

 
   - Ordinal: ordered rank (natural order is present)
        - Letter grade, customer satisfaction (very satisfied...very dissatisfied)
                 

 
   - Categories are the same or different

When working with categorical data, it is usually most appropriate to visualize through 

- bar plots

- pie charts

## Histograms

Chart to display the *distribution* of numerical values using bins

Distribution - the frequency of values

* Uses the area principle
* The **area** of each bar is proportional to the percent of individuals in the corresponding bin

Useful for:
 - summarizing numerical data that are measured on an interval scale - continuous data
 - To illustrate the major features of the distribution of the data in a convenient form.
 - When dealing with large data sets 

### Binning Numerical Values

Binning is counting the number of numerical values that lie within ranges called bins

* Bins are defined by their lower bounds (inclusive)
* The upper bound is the lower bound of the next bin

## Histogram in Code

````python
actors.hist('Number of Movies')
plt.show()
````

Default is 10 equal sized bins


But, we can create our own bins! (Be careful with this power!!!)

````python
actors.hist('Number of Movies', bins = 20) #default bins is 10

### additional formating
plt.ylabel("Frequency")
plt.xlabel("Number of Movies")
plt.title("Number of Movies from 50 Actors")
plt.show()
````

````python
actors.hist('Number of Movies', bins = 4) #default bins is 10

### additional formating
plt.ylabel("Frequency")
plt.xlabel("Number of Movies")
plt.title("Number of Movies from 50 Actors")
plt.show()
````

````python
my_bins = np.array([0, 20, 40, 60, 80])
actors.hist('Number of Movies', bins = my_bins)

### additional formating
plt.ylabel("Frequency")
plt.xlabel("Number of Movies")
plt.title("Number of Movies from 50 Actors")
plt.show()
````

Smaller bins can show some interesting detials

Finer-grained view of the distribution, but may not be as smooth of a distribution/histogram

## Bar Charts

Can display the *distribution* of a categorical variable (ie movie studios)

* One bar for each category
* Length of bar is some quantity
* You can choose the order of the bars

Compare quantity across categories:
* % of smartphone owners who have used their phone for the following in the last year: online banking, job search, etc
* Gross ticket sales for individual movies

````python
actors.plot.bar(x='Actor', y ='Average per Movie') #again we specify x and y axis
plt.show()
````

````python
actors.sort_values(by = 'Average per Movie', ascending = False)[0:10].plot.bar(x='Actor',
                                                       y ='Average per Movie') 
                                            #again we specify x and y axis
plt.show()
````

````python
actors.sort_values(by = 'Average per Movie', ascending = False).head(10).plot.bar(x='Actor',
                                                       y ='Average per Movie') 
                                            #again we specify x and y axis
plt.show()
````

````python
actors.sort_values(by = 'Average per Movie', ascending = False)[0:10].plot.barh(x='Actor',
                                                       y ='Average per Movie', legend = False) 
                                            #again we specify x and y axis
    
plt.title('Top Ten Actors by Average Gross Movies')  
plt.xlabel('Average per movie')

plt.show()
````

This is meaningful! But I'm curious how this compares to each of these actor's total gross. 

Let's graph that.

### How do we start?

````python
top_ten = actors.sort_values(by = 'Average per Movie', ascending = False)[:10]
top_ten
````

````python
top_ten.plot.barh(x='Actor', y ='Total Gross', legend = False)
plt.show()
````

## Can I plot these on the same graph?!

How?

We can manipulate the DataFrame and then graph it!

What does this mean?

 - condition based extraction
 - location methods
 - groupby
 -  pivot_table

````python
smaller_df = top_ten.set_index('Actor')[['Average per Movie', 'Total Gross']] 
smaller_df.head()
````

````python
### this line of code does the same thing!!!!
top_ten.pivot_table(['Average per Movie', 'Total Gross'], 
                    index= 'Actor', aggfunc='sum')
````

````python
smaller_df.plot.barh(xlabel= 'Money', title = 'Top 10 Largest Average Income per Movie Against Total Gross', figsize=(10, 12))


plt.legend(loc='upper right')
plt.show()
````

## Pie Charts

A popular way of visualizing categorical data

Follow the "Area Principle" - the area should be proportional to the value being represented

Have fallen out of fashion with some data scientists...

````python
sample_class_data = pd.DataFrame(
    {   'Fav food'  :["Classic Wisconsin Butter Burger",
                             "Pizza", 'Sushi', "I’m a big 'grilled cheese sandwich' person.",
                            'Pasta'],
     'Counts': [1,4,3,1,5]
   })  
sample_class_data.set_index('Fav food', inplace = True)
````

````python
sample_class_data.plot.pie(y='Counts',label = '',
                   title = 'Fav foods', 
                   figsize=(10, 10), autopct='%1.1f%%', legend = False)
## Note: we've added the argument autopct='%1.1f%%' to add percents
plt.show()
````

## How to make 'good' graphs:


* Type of graph matches the type of data being graphed

* Clear axis labels for x and y

* Does not violate the area principle

* Shows the full picture

**Summary:**
    
`Scatter plot`: relation between numerical variables

`Line graph`: sequential data (over time, etc.)

`Histogram`: distribution of numerical data

`Bar chart`: distribution of categorical data


`Pie Chart`: distribution of categorical data (out of fashion in some data visualization circles)
